```python
# Importing the Air traffic file from the local directory
path=os.getcwd()
airTraffic_file_path=path+"\\Air_Traffic_Passenger_Statistics.csv"
# creating a dataframe for the Air Traffic file 
airTraffic_orig_df=pd.read_csv(airTraffic_file_path)
# Printing top 5 rows from Air Traffic Dataframe
airTraffic_orig_df.head()
```

```python
# Taking a copy  of the dataframe
airTraffic_df1=airTraffic_orig_df.copy()
```

```python
# Getting the number of columns and Rows in the  Air Traffic  dataframe
airTraffic_columns=[col for col in airTraffic_df1.columns]
print(f"There are {len(airTraffic_columns)} Columns and {len(airTraffic_df1)} Rows in Air Traffic Dataframe")
```

```python
# Printing the shape of the dataframe before and after cleansing
print(f"Shape Before Removing Dupes : {airTraffic_df1.shape}")
# In place operation is performed by removing the duplicates
airTraffic_df1.drop_duplicates(keep='first',inplace=True)
print(f"Shape After Removing Dupes : {airTraffic_df1.shape}")
```

```python
# Checking for nulls in the columns.The Nulls are only present in the  Airline codes and it will be 
# ignored, as those columns are not used for the research
airTraffic_df1.isnull().sum()
```

```python
# Creating a data column from Activity period column
airTraffic_df1["Activity_date"] = pd.to_datetime(airTraffic_df1["Activity Period"], format='%Y%m')
# Creating new columns for Year and Month
airTraffic_df1["Activity_year"]=airTraffic_df1["Activity_date"].dt.year
airTraffic_df1["Activity_month"]=airTraffic_df1["Activity_date"].dt.month
```

#### 4.1: Creating a Line plot of Year vs Passenger count

```python
# Creating new dataframe by grouping the data based on Year and getting sum of passengers
airTraffic_df2=airTraffic_df1.groupby("Activity_year").sum()["Passenger Count"].reset_index()
# Excluding the years 2005 and 2023 from the results as we only have partial data for those years
airTraffic_df2=airTraffic_df2[~airTraffic_df2["Activity_year"].isin([2005,2023])]
# Printing top 5 rows from the grouped dataframe
airTraffic_df2.head()
```

```python
# PLOT 1: Creating a Line plot of Year vs Passenger count
plt.plot(airTraffic_df2["Activity_year"],airTraffic_df2["Passenger Count"]/1000000,label="Passengers",linestyle='--',marker="*",color='#7B27A8')
# Setting axis limits and labels
plt.xticks(rotation=45)
plt.ylim(0,60)
plt.title("Passengers count at SFO Airport (Plot #1)")
plt.ylabel("Passengers (in Millions)")
plt.tight_layout()
```

##### ***Interpretation of the plot1:*** Plot 1  indicates that 2018 recorded the most number of travellers which was closely followed by 2019. In 2020, as expected the air travel plunged due to the pandemic, but the numbers had recovered significantly as of 2022. As we only have partial data for 2023, the data was not included in the research.

#### 4.2: Creating a Bar plot of Months of the Year vs Passenger count

```python
# Filtering the Air Traffic data for the years 2007, 2013 and 2019
airTraffic_df3=airTraffic_df1[airTraffic_df1["Activity_year"].isin([2007,2013,2019])]
# Creating new dataframe by grouping the data based on Year and getting sum of passengers
airTraffic_df4=airTraffic_df3.groupby(["Activity_year","Activity_month"]).sum()["Passenger Count"].reset_index()
airTraffic_df4.head()
```

```python
# Creating dataframes for each year to represent them in the Bar plot
airTraffic_df4_2007=airTraffic_df4[airTraffic_df4["Activity_year"].isin([2007])]
airTraffic_df4_2013=airTraffic_df4[airTraffic_df4["Activity_year"].isin([2013])]
airTraffic_df4_2019=airTraffic_df4[airTraffic_df4["Activity_year"].isin([2019])]
# Creating x indexes with same number of elements as Year. This will be used in the plotting of x axis
x_indexes=np.arange(len(airTraffic_df4_2019["Activity_year"]))
# Creating a list of months which will be used as labels for Xaxis
x_months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
# PLOT 2: Creating a Bar plot of Months of the Year vs Passenger count
# Assigning default width for each bar in the Barchart
width=0.25
# Plotting BarChart for each year. The Value of the Xindexes are adjusted by using the Width 
plt.bar(x_indexes-width,airTraffic_df4_2007["Passenger Count"]/1000000,width=width,color="#e5ae37",label="2007")
plt.bar(x_indexes,airTraffic_df4_2013["Passenger Count"]/1000000,width=width,color="#7D439C",label="2013")
# The xindexes are adjusted according to the width of the Bar chart
plt.bar(x_indexes+width,airTraffic_df4_2019["Passenger Count"]/1000000,width=width,color="#759C43",label="2019")
plt.legend()
# The x-ticks will have values of x_indexes with label values as Month.
plt.xticks(ticks=x_indexes,labels=x_months)
# Setting the Title, X and Y labels 
plt.legend()
plt.xticks(rotation=45)
plt.title("Passenger Count at SFO Airport (Plot #2)")
plt.ylabel("Passengers (in Millions)")
plt.tight_layout()
plt.show()
```

##### ***Interpretation of the plot2:*** The Barplot results indicate that August was the busiest month followed by July and June. Analysis was done on the years 2007, 2013 and 2019(each 6 years apart) to study the pattern and they all show similar results.  Febuary recorded the fewest number of travellers compared to rest of the year.

#### 4.3:  Creating a Line plot of Months of the Year vs Passenger count for each type of Travel

```python
# Creating new dataframe by grouping the data based on Year, Geo Summary and getting sum of passengers
airTraffic_df5=airTraffic_df1.groupby(["Activity_year","GEO Summary"]).sum()["Passenger Count"].reset_index()
# Exclusing the years 2005 and 2023 as they have partial data
airTraffic_df5=airTraffic_df5[~airTraffic_df5["Activity_year"].isin([2005,2023])]
# Printing top 5 rows of the dataframe
airTraffic_df5.head()
```

```python
# Creating dataframes for each type to represent them in the line plot
airTraffic_df5_D=airTraffic_df5[airTraffic_df5["GEO Summary"].isin(["Domestic"])]
airTraffic_df5_I=airTraffic_df5[airTraffic_df5["GEO Summary"].isin(["International"])]
```
